#libraries.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#Experiment 1

##Applying Horizontal, Vertical, and Diagonal Sobel Filters.

```python
def apply_sobel_filters(image_path):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print("Error: Image not found!")
        return

    # Sobel kernels
    sobel_horizontal = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
    sobel_vertical = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_diagonal = np.array([[0, 1, 2], [-1, 0, 1], [-2, -1, 0]])

    # Apply Sobel filters using cv2.filter2D
    horizontal_edges = cv2.filter2D(image, -1, sobel_horizontal)
    vertical_edges = cv2.filter2D(image, -1, sobel_vertical)
    diagonal_edges = cv2.filter2D(image, -1, sobel_diagonal)

    # Display results
    plt.figure(figsize=(10, 7))

    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.imshow(horizontal_edges, cmap='gray')
    plt.title('Horizontal Sobel')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.imshow(vertical_edges, cmap='gray')
    plt.title('Vertical Sobel')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.imshow(diagonal_edges, cmap='gray')
    plt.title('Diagonal Sobel')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
image_path ='/content/drive/MyDrive/Ateacherimage.jpg'
apply_sobel_filters(image_path)
```

#Experiment 3

##Applying Laplacian Filter for Edge Detection.

```python
def apply_laplacian_filter(image_path):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print("Error: Image not found!")
        return

    # Apply Laplacian filter
    laplacian_edges = cv2.Laplacian(image, cv2.CV_64F)
    laplacian_edges = np.uint8(np.absolute(laplacian_edges))

    # Display results
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(laplacian_edges, cmap='gray')
    plt.title('Laplacian Filter')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
image_path = '/content/drive/MyDrive/Ateacherimage.jpg'
apply_laplacian_filter(image_path)
```

#Our analysis

##1. Sobel Operator

**How do horizontal, vertical, and diagonal Sobel filters emphasize edges differently?**

- Horizontal Sobel filters detect edges where there are significant intensity changes along the vertical axis, emphasizing horizontal structures in the image, such as lines or patterns running left to right.
- Vertical Sobel filters detect edges where there are significant intensity changes along the horizontal axis, highlighting vertical structures like lines or patterns running up and down.
- Diagonal Sobel filters, using custom diagonal kernels, are sensitive to edges running diagonally, detecting intensity changes along slanted lines and capturing structures not aligned with primary horizontal or vertical directions.

**What happens to the diagonal edges when you increase the Sobel kernel size?**

- Edge Sensitivity: Smaller kernels detect sharp edges and finer details but are prone to noise, while larger kernels reduce noise and smooth the image, potentially losing finer edges.

- Edge Thickness: Smaller kernels produce thinner edges for precise localization, whereas larger kernels yield thicker edges for a more generalized representation.

- Computational Cost: Smaller kernels are computationally efficient, while larger kernels increase computational load due to involving more pixels in convolution.

- Scale of Edges Detected: Smaller kernels are effective for detecting small features, while larger kernels capture broader structures and trends in the image.

- Parameter Tuning: Kernel size selection depends on the application and image characteristics, requiring empirical testing to determine optimal performance.

##2. Prewitt Operator

**Compare the results of Prewitt and Sobel operators. Which one is better for noise suppression and why?**

- The Sobel operator is better for noise suppression than the Prewitt operator because it uses Gaussian smoothing and differentiation to create stronger noise resistance.
-The Sobel operator also provides better edge orientation information than the Prewitt operator.
-The Sobel operator calculates the weighted difference in intensity of pixels around a given pixel. The Prewitt operator applies an approximation to the derivatives of the intensity function to calculate their edges.
-The Sobel operator emphasizes pixels closer to the center of the mask, while the Prewitt operator does not.

##3. Laplacian Filter

**How does the Laplacian operator differ from Sobel and Prewitt?**

- The Laplacian filter works on the second derivative of an image, while the Sobel and Prewitt filters work on the first derivative.
-The Laplacian filter uses a single symmetrical kernel, while the Sobel filter uses horizontal and vertical kernels.
-The Laplacian filter is sensitive to finer details, while the Sobel filter has a smoothing effect.
-The Laplacian filter is sensitive to noise, so the image is often Gaussian smoothed before applying the filter.


**Explain the importance of second-order derivatives in edge detection.**

- Second-order derivatives in edge detection are crucial because they measure the rate of change of the gradient, allowing for the identification of rapid intensity variations in an image.
- These derivatives highlight regions where intensity changes abruptly, making them effective for detecting edges.
- They also provide more precise localization of edges by identifying zero-crossings, where the gradient transitions from positive to negative or vice versa.
- This property is particularly useful in emphasizing fine details and enhancing features that might be missed by first-order methods, although it comes at the cost of increased sensitivity to noise.

**Which filter (Sobel, Prewitt, or Laplacian) would you choose for detecting faint edges and why?**

- The Laplacian filter is the best choice for detecting faint edges because it uses second-order derivatives to emphasize regions of rapid intensity change, making it highly sensitive to subtle edges.
- Unlike Sobel and Prewitt, which rely on directional gradients, the Laplacian is direction-independent and captures edges regardless of orientation.
- This property ensures that faint edges, which may not align with standard horizontal or vertical directions, are also detected. However, due to its sensitivity to noise, preprocessing steps like smoothing or denoising may be necessary to achieve optimal results.

**Apply these filters on a noisy image. Which one performs best?**

- When applied to a noisy image, the Sobel filter generally performs best for edge detection because its kernel design includes smoothing elements that help reduce the effect of noise while still detecting edges.
- The Prewitt filter is slightly less effective in noisy conditions due to its simpler kernel that lacks weighted smoothing.
- The Laplacian filter, while excellent for capturing faint edges, amplifies noise significantly due to its reliance on second-order derivatives, making it less robust without prior denoising steps.
